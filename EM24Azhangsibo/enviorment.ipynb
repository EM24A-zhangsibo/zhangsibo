```python
import pandas as pd
import matplotlib.pyplot as plt

# 导入温度数据
df = pd.read_csv(
    "https://data.giss.nasa.gov/gistemp/tabledata_v4/NH.Ts+dSST.csv",
    skiprows=1,
    na_values="***",
)

# 设置年份为索引
df = df.set_index("Year")

# 选择季节的列
seasons = ['DJF', 'MAM', 'JJA', 'SON']

for season in seasons:
    fig, ax = plt.subplots()
    ax.axhline(0, color="orange", linestyle='--', label='1951–1980 average')
    ax.annotate("1951—1980 average", xy=(0.8, 0), xycoords=("axes fraction", "data"), color='orange')

    df[season].plot(ax=ax)
    ax.set_title(f"Average temperature anomaly in {season} \n in the northern hemisphere (1880—{df.index.max()})")
    ax.set_ylabel("Temperature anomaly (degrees Celsius)")
    ax.set_xlabel("Year")
    ax.legend()
    
    # 添加结论
    conclusion_text = "显著的上升趋势，气候变暖正在加剧。"
    ax.text(0.05, 0.9, conclusion_text, transform=ax.transAxes, fontsize=12, 
            verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))

    plt.show()
```

```python
# 绘制年度平均气温异常折线图
fig, ax = plt.subplots()
ax.axhline(0, color="orange", linestyle='--', label='1951–1980 average')
ax.annotate("1951—1980 average", xy=(0.8, 0), xycoords=("axes fraction", "data"), color='orange')

df["J-D"].plot(ax=ax)  # "J-D"列为年度平均温度异常
ax.set_title(f"Average annual temperature anomaly \n in the northern hemisphere (1880—{df.index.max()})")
ax.set_ylabel("Temperature anomaly (degrees Celsius)")
ax.set_xlabel("Year")
ax.legend()

# 添加结论
annual_conclusion_text = "年度气温异常显示出上升趋势，全球变暖的迹象明显。"
ax.text(0.05, 0.9, annual_conclusion_text, transform=ax.transAxes, fontsize=12, 
        verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))

plt.show()
```

```python
# Create a variable that has years 1951 to 1980, and months Jan to Dec (inclusive)
temp_all_months = df.loc[(df.index >= 1951) & (df.index <= 1980), "Jan":"Dec"]
# Put all the data in stacked format and give the new columns sensible names
temp_all_months = (
    temp_all_months.stack()
    .reset_index()
    .rename(columns={"level_1": "month", 0: "values"})
)
# Take a look at this data:
temp_all_months
```

```python
quantiles = [0.3, 0.7]
list_of_percentiles = np.quantile(temp_all_months["values"], q=quantiles)


print(f"The cold threshold of {quantiles[0]*100}% is {list_of_percentiles[0]}")
print(f"The hot threshold of {quantiles[1]*100}% is {list_of_percentiles[1]}")
```

```python
# Create a variable that has years 1981 to 2010, and months Jan to Dec (inclusive)
temp_all_months = df.loc[(df.index >= 1981) & (df.index <= 2010), "Jan":"Dec"]
# Put all the data in stacked format and give the new columns sensible names
temp_all_months = (
    temp_all_months.stack()
    .reset_index()
    .rename(columns={"level_1": "month", 0: "values"})
)
# Take a look at the start of this data data:
temp_all_months.head()
```

```python
entries_less_than_q30 = temp_all_months["values"] < list_of_percentiles[0]
proportion_under_q30 = entries_less_than_q30.mean()
print(
    f"The proportion under {list_of_percentiles[0]} is {proportion_under_q30*100:.2f}%"
)
```

```python
proportion_over_q70 = (temp_all_months["values"] > list_of_percentiles[1]).mean()
print(f"The proportion over {list_of_percentiles[1]} is {proportion_over_q70*100:.2f}%")
```

```python
temp_all_months = (
    df.loc[:, "DJF":"SON"]
    .stack()
    .reset_index()
    .rename(columns={"level_1": "Season", 0: "Values"})
)
temp_all_months["Period"] = pd.cut(
    temp_all_months["Year"],
    bins=[1921, 1950, 1980, 2010],
    labels=["1921—1950", "1951—1980", "1981—2010"],
    ordered=True,
)
# Take a look at a cut of the data using `.iloc`, which provides position
temp_all_months.iloc[-135:-125]
```

```python
grp_mean_var = temp_all_months.groupby(["Season", "Period"])["Values"].agg(
    [np.mean, np.var]
)
grp_mean_var
```

```python
# 导入温度数据
df_temp = pd.read_csv(
    "https://data.giss.nasa.gov/gistemp/tabledata_v4/NH.Ts+dSST.csv",
    skiprows=1,
    na_values="***",
)

# 设置年份为索引
df_temp = df_temp.set_index("Year")

# 导入 CO2 数据
df_co2 = pd.read_csv("data/1_CO2-data.csv")

# 选择 CO2 的趋势数据，这里以六月（6月）和八月（8月）为例
df_co2_june = df_co2[df_co2["Month"] == 6][["Year", "Trend"]]
df_co2_august = df_co2[df_co2["Month"] == 8][["Year", "Trend"]]

# 合并 CO2 数据与温度数据（6月）
df_temp_co2_june = pd.merge(df_co2_june, df_temp, on="Year")

# 合并 CO2 数据与温度数据（8月）
df_temp_co2_august = pd.merge(df_co2_august, df_temp, on="Year")

# 绘制 六月 图表
plt.figure(figsize=(14, 6))

# 六月的数据
plt.subplot(1, 2, 1)
plt.scatter(df_temp_co2_june["Jun"], df_temp_co2_june["Trend"], color='blue', alpha=0.5)
plt.title("June Temperature Anomalies vs CO2 Levels")
plt.xlabel("Temperature Anomaly (June)")
plt.ylabel("CO2 Levels (Trend)")
plt.axhline(0, color="orange", linestyle='--', label='1951-1980 average')
plt.axvline(0, color="orange", linestyle='--')
plt.grid()
plt.legend()

# 计算相关系数
correlation_june = df_temp_co2_june["Jun"].corr(df_temp_co2_june["Trend"])
plt.text(0.05, 0.95, f"Correlation: {correlation_june:.2f}", 
         transform=plt.gca().transAxes, fontsize=12, color='black', 
         bbox=dict(facecolor='white', alpha=0.5))

# 添加结论 
plt.text(0.05, 0.85, 
         "六月的温度异常与CO2水平有显著的正相关。\n需要谨慎，不能单凭相关性得出因果关系。",
         transform=plt.gca().transAxes, fontsize=12, color='black', 
         bbox=dict(facecolor='lightyellow', alpha=0.5))

# 绘制 八月 图表
plt.subplot(1, 2, 2)
plt.scatter(df_temp_co2_august["Aug"], df_temp_co2_august["Trend"], color='green', alpha=0.5)
plt.title("August Temperature Anomalies vs CO2 Levels")
plt.xlabel("Temperature Anomaly (August)")
plt.ylabel("CO2 Levels (Trend)")
plt.axhline(0, color="orange", linestyle='--')
plt.axvline(0, color="orange", linestyle='--')
plt.grid()
plt.legend()

# 计算相关系数
correlation_august = df_temp_co2_august["Aug"].corr(df_temp_co2_august["Trend"])
plt.text(0.05, 0.95, f"Correlation: {correlation_august:.2f}", 
         transform=plt.gca().transAxes, fontsize=12, color='black', 
         bbox=dict(facecolor='white', alpha=0.5))

# 添加结论 
plt.text(0.05, 0.85, 
         "八月的温度异常与CO2水平也有明显的正相关。\n同样需要谨慎对待相关性与因果性。",
         transform=plt.gca().transAxes, fontsize=12, color='black', 
         bbox=dict(facecolor='lightyellow', alpha=0.5))

plt.tight_layout()
```
